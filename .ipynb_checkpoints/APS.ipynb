Seja f(x,y) = 3x^2 + x*y + y^2 + x + y

O vetor gradiente é dado por:

Deltaf = dfdx(x0,y0) + dfdy(x0,y0)

```python
def f(x,y):
    f = 3*x**2 + x*y + y**2 + x + y
    return f
    
#Calculando o vetor gradiente:
def gradiente(x,y):
    dfdx = 6*x + y +1
    dfdy = x + 2*y + 1 
    return (dfdx, dfdy)

#Para achar novos pontos x1, y1:
def pontos(x1, y1, alpha):
    x = 0
    y = 0
    alpha = 0.1
    x1 = x - alpha*dfdx
    y1 = y - alpha*dfdy
    return(x1,y1)
           
print(pontos(x1,y1,alpha))
```
