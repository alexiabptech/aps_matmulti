```python
#Primeiro, precisamos dizer ao programa qual é a nossa função em análise:
def f(x,y):
    f = 3*x**2 + x*y + y**2 + x + y
    return f

#Depois, calcula-se  o vetor gradiente, para conseguirmos entender qual é o ponto de mínimo que a etapa 1 requer:
def gradiente(x,y):
    dfdx = 6*x + y +1
    dfdy = x + 2*y + 1 
    return (dfdx, dfdy)

#Para achar novos pontos x1, y1, utilizamos o sistema dado pela orientação, no qual calculamos novos pontos em função do vetor gradiente de um passo
#Esse passo seria o "avanço" ou "regresso" do vetor:
def passo(x0, y0, alpha):
    a = gradiente(x0,y0)[0] #Avanço em x
    x1 = x0 - a*alpha
    
    b = gradiente(x0,y0)[1] #Avanço em y
    y1 = y0 - b*alpha
    return(x1,y1)

x0 = 0
y0 = 0
alpha = 0.1
print(passo(x0,y0,alpha)) #print calcula o ponto de mínimo da função f(x,y)

```

## D) Variando valores de alpha, com novas iterações

```python
def f(x, y):
    return 3 * x ** 2 + x * y + y ** 2 + x + y

def gradiente(x, y):
    dfdx = 6 * x + y + 1
    dfdy = x + 2 * y + 1
    return dfdx, dfdy

def passo(x, y, alpha):
    dfdx, dfdy = gradiente(x, y)
    xf = x - dfdx * alpha
    yf = y - dfdy * alpha
    return xf, yf

x = 0
y = 0
alphas = [0.1, 0.15, 0.2, 0.3, 0.5]

for i, alpha in enumerate(alphas):
    x, y = passo(x, y, alpha)
    print(f"x{i+1}: {x}")
    print(f"y{i+1}: {y}")
```

## Etapa 2 - Determinando os pontos de mínimo da função g(x,y):


```python
import math
def calcula_gradiente(x, y):
    df_dx = ((x*((x**2 + y**2 + 3)**(-0.5))) + (2*x*(math.exp(-y**2))) + 2*x -6)
    df_dy = ((y*((x**2 + y**2 + 3)**(-0.5))) - (2*y*(x**2)*(math.exp(-y**2))))
    return [df_dx, df_dy]

def pega_ponto(x0, y0, passo):
    x = y0
    y = x0
    #conta o numero de iterações para chegar no mínimo
    cont_iteracao = 0
    state = True
    while state == True:
        gradiente = calcula_gradiente(x, y)
        if math.sqrt(gradiente[0]**2 + gradiente[1]**2) <= 1e-5:
            state = False
        x -= passo*gradiente[0]
        y -= passo*gradiente[1]
        cont_iteracao += 1
    return x,y,cont_iteracao

# primeiro ponto mínimo de g
x1 = - 3
y1 = - 3

x_1min ,y_1min, iteracoes1 = pega_ponto(x1, y1,0.1)

print("O primeiro ponto é ({:.5f},{:.5f}) e o numero iterações {}".format(x_1min,y_1min,iteracoes1))

# Procurando o segundo ponto mínimo da função g
x2 = 3
y2 = 3

x_2min ,y_2min, iteracoes2 = pega_ponto(x2, y2,0.1)

print("O segundo ponto é ({:.5f},{:.5f}) e o numero iterações {}".format(x_2min,y_2min,iteracoes2))

# para diferentes passos: 
passos = [0.15, 0.2, 0.3, 0.5]

i1 = 0
while i1 < len(passos):
    x1 = - 3
    y1 = - 3
    x_1min ,y_1min, iteracoes1 = pega_ponto(x1, y1,passos[i1])

    print("O primeiro ponto é ({:.5f},{:.5f}) e o numero iterações {}, para o passo : {}".format(x_1min,y_1min,iteracoes1,passos[i1]))
    i1 += 1
    
i2 = 0
while i2 < len(passos):
    x2 = 3
    y2 = 3
    x_2min ,y_2min, iteracoes2 = pega_ponto(x2, y2,passos[i2])
    print("O segundo ponto é ({:.5f},{:.5f}) e o numero iterações {}, para o passo : {}".format(x_2min,y_2min,iteracoes2, passos[i2]))
    i2 += 1
```
